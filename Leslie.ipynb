# Funcion de Leslie

 - L = Matriz de Leslie
 - T = tiempo
 - I = Matriz Inicial
 
 $$I^{(T)}=L^{(T)}I$$

```python
def Crec(L,I,t):
    L=np.linalg.matrix_power(L,t)
    A=np.dot(L,I)
    return A
    
```

```python
Inicial=[144,60,20] #Cantidad de peces iniciales '[T_1,T_2,T_3]'
M=np.empty((11,3)) #Matriz vacia para almacenar los datos obtenidos
```

```python
for t in range(0,11):
    M[t]=Crec(les,Inicial,t)
    print(Crec(les,Inicial,t))
```

```python
M=M.round(decimals=6) #Acotador
Tipo1=M[0:,0]
Tipo2=M[0:,1]
Tipo3=M[0:,2]
```

# Autovalores

```python
evals= la.eigvals(les)
```

```python
Q=np.empty((30,3))
for t in range(0,30):
    Q[t]=Crec(les,Inicial,t)
    print(Q[t],t)
```

```python
TipoA1=Q[:,0]
TipoA2=Q[:,1]
TipoA3=Q[:,2]

Largo_Plazo_1=np.zeros(shape=(29,1))
Largo_Plazo_2=np.empty((29,1))
Largo_Plazo_3=np.empty((29,1))

np.size(TipoA1)
```

```python
np.size(Q)
```

```python
for t in range(1,29):
    Largo_Plazo_1[t]=(TipoA1[t])/(mt.pow(evals[0],t))
    Largo_Plazo_2[t]=(TipoA2[t])/(mt.pow(evals[1],t))
    Largo_Plazo_3[t]=(TipoA3[t])/(mt.pow(evals[2],t))
```
